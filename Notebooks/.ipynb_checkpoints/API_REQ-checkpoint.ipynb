```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime
import time

# Constants
R = 287.05  # Specific gas constant for dry air in J/(kg·K)
API_URL = "https://api.open-meteo.com/v1/forecast"

# Wind turbine parameters (adjustable)
TURBINE_RADIUS = 50  # Radius of the wind turbine blades in meters
EFFICIENCY_WIND = 0.4  # Efficiency of the wind turbine (40%)

# Solar panel parameters (adjustable)
EFFICIENCY_SOLAR = 0.2  # Efficiency of the solar panels (20%)
SOLAR_CONSTANT = 1361  # Solar constant in W/m² (average solar radiation at the top of the atmosphere)

def fetch_weather_data(latitude, longitude):
    """
    Fetches hourly weather data from the Open-Meteo API, including solar radiation.
    """
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "hourly": "temperature_2m,pressure_msl,wind_speed_10m,is_day,shortwave_radiation",
        "timezone": "auto"
    }
    try:
        response = requests.get(API_URL, params=params)
        response.raise_for_status()  # Raise an error for bad responses (4xx or 5xx)
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None


# Calculate air density
def calculate_density(pressure_pa, temperature_k):
    try:
        return pressure_pa / (R * temperature_k)
    except (TypeError, ZeroDivisionError):
        return np.nan


def calculate_wind_power(wind_speed_kmh, air_density, turbine_radius=TURBINE_RADIUS, efficiency=EFFICIENCY_WIND):
    """
    Calculates wind power generated using the wind power formula.

    Args:
        wind_speed_kmh (float): Wind speed in km/h.
        air_density (float): Air density in kg/m³.
        turbine_radius (float): Radius of the wind turbine blades in meters.
        efficiency (float): Efficiency of the wind turbine.

    Returns:
        float: Wind power generated in watts.
    """
    wind_speed_ms = wind_speed_kmh * (1000 / 3600)  # Convert km/h to m/s
    swept_area = np.pi * turbine_radius**2  # A = πR²
    power = 0.5 * air_density * swept_area * wind_speed_ms**3 * efficiency
    return power

def calculate_solar_power(is_day, solar_radiation, efficiency=EFFICIENCY_SOLAR):
    """
    Calculates solar power generated based on solar radiation and daylight.

    Args:
        is_day (int): 1 if daytime, 0 if nighttime.
        solar_radiation (float): Solar radiation in W/m².
        efficiency (float): Efficiency of the solar panels.

    Returns:
        float: Solar power generated in watts.
    """
    if is_day == 0:  # Nighttime
        return 0
    return solar_radiation * efficiency

def process_weather_data(data):
    if not data or "hourly" not in data:
        print("No hourly data found in the API response.")
        return None

    hourly_data = data["hourly"]
    
    # Ensure no null values cause errors
    wind_speeds = hourly_data.get("wind_speed_10m", [])
    wind_speeds = [ws/3.6 if ws is not None else np.nan for ws in wind_speeds]
    
    pressures = hourly_data.get("pressure_msl", [])
    pressures = [(p * 100)*0.0002952998057228 if p is not None else np.nan for p in pressures]  # Convert hPa to InHg
    
    temperatures = hourly_data.get("temperature_2m", [])
    temperatures = [t if t is not None else np.nan for t in temperatures]
    
    solar_radiation = hourly_data.get("shortwave_radiation", [])
    solar_radiation = [sr if sr is not None else np.nan for sr in solar_radiation]
    
    is_day = hourly_data.get("is_day", [])
    is_day = [day if day is not None else 0 for day in is_day]
    
    df = pd.DataFrame({
        "Time": hourly_data.get("time", []),
        "Is_Day": is_day,
        "Average_Wind_Speed (m/s)": [ws for ws in wind_speeds],
        "Average_Barometric_Pressure (InHg)": pressures,
        "Temperature (K)": [t + 273.15 for t in temperatures],
        "Solar_Radiation (W/m²)": solar_radiation
    })

    df["Density (kg/m³)"] = [
        calculate_density(p/(100*0.0002952998), t) for p, t in zip(df["Average_Barometric_Pressure (InHg)"], df["Temperature (K)"])
    ]

    df["Wind_Power_Generated (W)"] = df.apply(
        lambda row: calculate_wind_power(row["Average_Wind_Speed (m/s)"], row["Density (kg/m³)"]), axis=1
    )

    df["Solar_Power_Generated (W)"] = df.apply(
        lambda row: calculate_solar_power(row["Is_Day"], row["Solar_Radiation (W/m²)"]), axis=1
    )

    return df

def save_data_to_csv(df, filename):
    """
    Saves the DataFrame to a CSV file.
    """
    df.to_csv(filename, index=False)
    print(f"Data saved to {filename}")

def run_weekly_data_fetch(latitude, longitude, filename="weather_data.csv"):
    """
    Runs the data fetch process once a week. Retries indefinitely until data is successfully fetched.
    """
    while True:
        print("Attempting to fetch weather data...")
        data = fetch_weather_data(latitude, longitude)
        if data:
            df = process_weather_data(data)
            if df is not None:
                save_data_to_csv(df, filename)
                print("Data fetch and processing successful.")
                break  # Exit the loop if data is successfully fetched and processed
        else:
            print("Failed to fetch data. Retrying in 1 hour...")
            time.sleep(3600)  # Wait for 1 hour before retrying


def main():
    """
    Main function to run the script.
    """
    try:
        # Get user input for latitude and longitude
        latitude = float(input("Enter latitude: "))
        longitude = float(input("Enter longitude: "))

        # Fetch weather data
        print("Fetching weather data...")
        weather_data = fetch_weather_data(latitude, longitude)

        if weather_data:
            # Process data
            print("Processing data...")
            processed_data = process_weather_data(weather_data)

            if processed_data is not None:
                # Save data to CSV
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
                filename = f"enhanced_weather_data_{timestamp}.csv"
                save_data_to_csv(processed_data, filename)
    except ValueError:
        print("Invalid input. Please enter numeric values for latitude and longitude.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```
